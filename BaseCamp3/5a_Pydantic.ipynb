```python
from pydantic import BaseModel, ConfigDict
```

```python
# Define a Class with your specific data constraints
# Its derived from BaseModel and indicates the type of each
class DataSet (BaseModel) :

    id : str
    val_1 : int
    val_2 : int
    threshold : float
    output : float = 0.0
```

```python
# Define Objects based on the class that is defined.
# Try to change the types
d1 = DataSet (id='d1', val_1 = 2, val_2 = 3, threshold=4.5, output=2.8)
d2 = DataSet (id='d2', val_1 = 7, val_2 = 300, threshold=11.2)

print (d1)
print (d2)
```

```python
# Use as Object 
print (d1.id)
print (d1.val_1)
```

```python
# Use the methods for JSON, Dict
print (d1.model_dump_json())
v1 = d1.model_dump ()
v1
d1.model_validate_json
```

```python
# Make strict typing ...

class DataSet (BaseModel) :

    model_config = ConfigDict(strict=True)

    id : str
    val_1 : int
    val_2 : int
    threshold : float
    output : float = 0.0
```

```python
d1 = DataSet (id='d1', val_1 = 2, val_2 = '3', threshold=4.5, output=2.8)

print (d1)
```
